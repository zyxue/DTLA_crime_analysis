# Some quick stats

```python
adf.head()
```

### Crime by year

```python
ax = adf.groupby(['year', 'offense_type']).size().unstack().plot.bar(stacked=True, edgecolor='white')
ax.legend(bbox_to_anchor=(1.01, 1), loc='upper left')
ax.grid()
ax.set_ylabel('# crimes')
```

So the crime number is actually increasing in recent years, but it could be due to better way of reporting crimes.

### Crime by month

```python
ax = adf.groupby(['month', 'offense_type']).size().unstack().plot.bar(stacked=True, edgecolor='white')
ax.legend(bbox_to_anchor=(1.01, 1), loc='upper left')
ax.grid()
ax.set_ylabel('# crimes')
```

Not much difference from month to month

### Crime by day

```python
ax = adf.groupby(['day', 'offense_type']).size().unstack().plot.bar(stacked=True, edgecolor='white')
ax.legend(bbox_to_anchor=(1.01, 1), loc='upper left')
ax.grid()
ax.set_ylabel('# crimes')
```

Not much difference from day to day

### Crime by day of the week

```python
ax = adf.groupby(['day_of_week', 'offense_type']).size().unstack().plot.bar(stacked=True, edgecolor='white')
ax.legend(bbox_to_anchor=(1.01, 1), loc='upper left')
ax.grid()
ax.set_ylabel('# crimes')
```

0 is Monday ([ref](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.dt.dayofweek.html)). So Friday and Saturday are slightly worse.

### Crime by hour

```python
ax = adf.groupby(['hour', 'offense_type']).size().unstack().plot.bar(stacked=True, edgecolor='white')
ax.legend(bbox_to_anchor=(1.01, 1), loc='upper left')
ax.grid()
ax.set_ylabel('# crimes')
```

Crime number goes up quickly and remains high during the day
